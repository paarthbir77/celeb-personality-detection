```python
print("running")
celeb = ["Kartik","Salman","Aamir","Saif","Hrithik","Akshay","Ajay","Deepika","Aishwarya","Priyanka"]
descriptors = ["HE","SHE","THE ACTOR","THE SINGER","THE DIRECTOR","THE DANCER","THE PRODUCER"]
for person in celeb:
    os.mkdir(("images/"+str(person)))
```

```python
def proc(text):
    text = re.sub('\n','.',text)
    text = re.sub('\r','.',text)
    text = re.sub('\t','.',text)
    text = re.sub('{','.',text)
    text = re.sub('}','.',text)
    text = re.sub('\n','.',text)
    text = re.sub('.com','.',text)
    text = re.sub('html','.',text)
    text = re.sub('.com','.',text)
    text = re.sub("[/|{}»]",'.',text)
    text = re.sub("[^a-zA-z0-9 ,#.-]","",text)
    return text
```

```python
def getinfo(url, celeb):
    print(url)
    res = requests.get(url)
    html = res.content
    hlink = res.text
    soup = bs(hlink, 'html5lib')
    links = str(soup.findAll('img'))
    celebinurl = []
    for person in celeb:
        if person.lower() in url:
            celebinurl.append(person)
    print(celebinurl)
    with open("textfiles/file.txt","a+") as text_file2:
        for sentence in links.split("<"):
            sentence = re.sub(r'"'," ", sentence)
            for word in sentence.split(" "):
                if word in celebinurl:
                    print(sentence, file=text_file2)
        text_file2.close()

    cont = bs(html,'html.parser')
    text = cont.get_text().strip()
    text = proc(text)
    #print(text)
    for sentence in text.split("."):
        for word in sentence.split(" "):
            if word in celebinurl:
                path = "celebrities/"+str(word)+".txt"
                with open(path, "a+") as text_file:
                    print(sentence, file=text_file)
```

```python
def getimageurl(file,celeb):
    text=""
    with open(file, "r") as text_file:
        text+= str(text_file.read())
    text_file.close()
    for sentence in text.split(">"):
        for word in sentence.split(" "):
            if word in celeb:
                path = "images/"+str(word)+"/"+str(word)+".txt"
                for wordurl in sentence.split(" "):
                    if "http" in str(wordurl):
                        #print(wordurl)
                        with open(path, "a+") as text_file:
                            print(wordurl,file=text_file)
                        text_file.close()
```

```python
def downloadimage():
    path = "images/"
    for dir in os.listdir(path):
        filepath = path+str(dir)
        if len(os.listdir(filepath))>0:
            #print(filepath)
            txtpath = filepath+"/"+filepath.split("/")[1]+".txt"
            #print(txtpath)
            text = ""
            with open(txtpath, "r") as text_file:
                text+= str(text_file.read())
                text_file.close()
            imagepath = filepath+"/"
            print(imagepath)
            sentences = []
            for sentence in text.split("\n"):
                #print(str(sentence))
                sentences.append(sentence.replace("\n",""))
            unique = set(sentences)
            unique.remove("")
            for item in unique:
                name = re.sub("https://","", str(item))
                name = re.sub("www","", name)
                name = re.sub("[/._;@!#$%&*,?]","", name)
                print(item)
                try:
                    with open(imagepath+name+".jpg","wb") as image_file:
                        image_file.write(requests.get(item).content)
                        image_file.close()
                except:
                    pass
```

```python
def geturls(person, url):
    page = requests.get(url)
    soup = bs(page.content)
    links = soup.findAll("a")
    text = str(links)
    text = re.sub(r"&",r"?",text)
    text = re.sub("<",r"?",text)
    text = re.sub("=",r"?",text)
    text = re.sub(r"%",r"?",text)
    sentences = []
    for sentence in text.split(r"?"):
        sentences.append(sentence.replace(r"\n", ""))
    unique = set(sentences)
    for item in unique:
        if person in item and "https" in item:
            with open("textfiles/links","a+") as text_file:
                print(item, file=text_file)
            text_file.close()

```
